### Exercise 37
**Task:** Create a plot with dual y-axis.

```python
# Your code here

```


````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)            
y2 = 100 * np.exp(-x)     
fig, ax1 = plt.subplots()
ax1.plot(x, y1, color='blue', label='sin(x)')
ax1.set_xlabel('X axis')
ax1.set_ylabel('sin(x)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red', label='100 * exp(-x)')
ax2.set_ylabel('100 * exp(-x)', color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.title("Dual Y-Axis Plot")
fig.tight_layout()  
plt.grid(True)
plt.show()
````

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
import numpy as np

t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, label='3D curve', color='purple')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Basic 3D Line Plot')

ax.legend()
plt.show()
````

### Exercise 42
**Task:** Use color maps in heatmaps.

```python
# Your code here

```


````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 10)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

cmaps = ['viridis', 'hot', 'coolwarm']

for ax, cmap in zip(axs, cmaps):
    im = ax.imshow(data, cmap=cmap)
    ax.set_title(f"Colormap: {cmap}")
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

plt.suptitle("Heatmaps with Different Colormaps")
plt.show()
````

````python
import matplotlib.pyplot as plt

sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['red','skyblue','green','orange','pink']

fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, pctdistance=0.85
)

centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

ax.axis('equal')  
plt.title("Donut Chart")
plt.show()
````

### Exercise 50
**Task:** Plot real-world data from a CSV file.

```python
# Your code here

```


````python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['2025-01-01', '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-05'],
    'Temperature': [23, 22, 21, 24, 23],
    'Humidity': [68, 70, 65, 67, 66]
}

df = pd.DataFrame(data)

csv_filename = 'weather_data.csv'
df.to_csv(csv_filename, index=False)
print(f"CSV file '{csv_filename}' created.")

df_loaded = pd.read_csv(csv_filename, parse_dates=['Date'])

df_loaded.plot(x='Date', y=['Temperature', 'Humidity'], marker='o', figsize=(8, 5))

plt.title("Temperature and Humidity Over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(True)
plt.show()
````
